```python
def calculator(x, op, y):
  if op == '+':
    z = x + y
  elif op == '-':
    z = x - y
  elif op == '*':
    z = x * y
  elif op == '/':
    z = x / y
  else:
    print('연산자가 잘못 입력됨')
    raise ValueError('연산자가 잘못 입력됨')
  return z

x = float(input('x의 값: '))
y = float(input('y의 값: '))

for i in range(1, 1000, 1):
  op = input('op: ')
  print(calculator(x, op, y))
```
