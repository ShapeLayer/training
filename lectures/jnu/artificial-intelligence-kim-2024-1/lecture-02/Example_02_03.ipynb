```python
import numpy as np
```

### 1차원 텐서
vector =  1차원 tensor
$$
x =
\begin{bmatrix}
1 \\
2
\end{bmatrix}
$$

### 2차원 텐서
$$
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$$

### 행렬곱
$$
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
1 \\
2
\end{bmatrix}
$$

### 3차원 텐서
$$
\begin{bmatrix}
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix} & \begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
\end{bmatrix}
$$

```python
x = np.array([1.0, 2.0, 3.0])
print(x)
print(x.shape)
```

```python
a = np.array(
  [
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
  ]
)
print(a)
print(a.ndim)
print(a.shape)
print(a.size)
print(a.data)
```

```python
b = np.array(
  [
    [[10, 11], [12, 13]],
    [[13, 12], [11, 10]]
  ]
)

print(a + b)
print(a - b)
print(a * b) # elemental wise (요소값끼리)
print(a @ b) # 행렬곱
```

### Broadcasting

```python
b * 10
```

## 매트플롯

```python
import matplotlib.pyplot as plt
x = np.arange(0, 6, .1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, label="cos")
plt.show()
```
