```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
X, y = datasets.load_diabetes(return_X_y=True)
X = X[:, 2]
```

```python
X
```

```python
W = 0  # weight, inclination
b = 0  # intercept

lrate = 0.01   # learning rate (alpha)
epochs = 1000  # loops

n = float(len(X))
```

```python
for i in range(epochs):
  y_pred = W*X + b  # prediction model

  # delta
  dW = (2 / n) * sum(X * (y_pred - y))
  db = (2 / n) * sum(y_pred - y)

  # next step
  W -= lrate * dW
  b -= lrate * db
```

```python
print(W, b)
y_pred = W * X + b
```

```python
# Scatter input data
plt.scatter(X, y)

# draw model (x-y function)
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')

# show
plt.show()
```
