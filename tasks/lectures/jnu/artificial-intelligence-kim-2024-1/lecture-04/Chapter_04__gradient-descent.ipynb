# Implementing Gradient Descent

## Initialize

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define Input

$ Vector(x, y) = ((0, 3), (1, 0), (2, 5.5))$

```python
# define input
X = np.array([0.0, 1.0, 2.0])
y = np.array([3.0, 3.5, 5.5])

# Scatter input data
plt.scatter(X, y)
```

### Define parameters, requirements

```python
W = 0  # weight, inclination
b = 0  # intercept

lrate = 0.01   # learning rate (alpha)
epochs = 1000  # loops

n = float(len(X))
```

## Define and Invoke Gradient Descent

```python
for i in range(epochs):
  y_pred = W*X + b  # prediction model

  # delta
  dW = (2 / n) * sum(X * (y_pred - y))
  db = (2 / n) * sum(y_pred - y)

  # next step
  W -= lrate * dW
  b -= lrate * db
```

## Show Result

```python
print(W, b)
y_pred = W * X + b
```

### Using Matplotlib

```python
# Scatter input data
plt.scatter(X, y)

# draw model (x-y function)
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')

# show
plt.show()
```
